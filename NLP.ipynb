```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize  import word_tokenize
```

```python
text= "i love natural Language proCessing!! yay its so much fun "
text=text.lower()
tokens=word_tokenize(text)
print(tokens)
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
#print(stop_words)
filtered_text= [word for word in tokens if word.lower() not in stop_words]
print(filtered_text)
```

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import wordnet
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng') # Download for POS tagging

stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()

#print(stemmer.stem('running')) # Stemming example
#print(lemmatizer.lemmatize('went',pos='v')) # Verb lemmatization example (went -> go)
#print(lemmatizer.lemmatize('better', pos='a')) # Adjective lemmatization example (better -> good)


def get_wordnet_pos(treebank_tag):
    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

stemmed_words = []
lemmatized_words_pos = []


tagged_tokens = nltk.pos_tag(filtered_text)
print(f"\nPOS Tagged Tokens:\n{tagged_tokens}\n")

for word, tag in tagged_tokens:

    stemmed_word = stemmer.stem(word)
    stemmed_words.append(stemmed_word)


    wn_tag = get_wordnet_pos(tag)
    lemmatized_word = lemmatizer.lemmatize(stemmed_word, pos=wn_tag)
    lemmatized_words_pos.append(lemmatized_word)

print(f"\nOriginal words (filtered):\n{filtered_text}\n")
print(f"\nStemmed words:\n{stemmed_words}\n")
print(f"\nLemmatized words (POS-aware):\n{lemmatized_words_pos}\n")
```

```python
# LEARNING BAG OF WORDS
from sklearn.feature_extraction.text import CountVectorizer
data = ["i love my job", "i love my pet", "i like my city"]
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data)

print("\nDocuments (Row Names):\n")
print(data)

print("\nVocabulary (Column Names/Feature Names):\n")
print(vectorizer.get_feature_names_out())

print("\nDocument-Term Matrix (X.toarray()):\n")
print(X.toarray())
```

```python
def preprocess_document(doc):
    doc = doc.lower()
    tokens = word_tokenize(doc)
    filtered_tokens = [word for word in tokens if word.lower() not in stop_words]
    tagged_tokens = nltk.pos_tag(filtered_tokens)
    processed_words = []
    for word, tag in tagged_tokens:
        stemmed_word = stemmer.stem(word)
        wn_tag = get_wordnet_pos(tag)
        lemmatized_word = lemmatizer.lemmatize(stemmed_word, pos=wn_tag)
        processed_words.append(lemmatized_word)
    return ' '.join(processed_words)

preprocessed_data = [preprocess_document(doc) for doc in data]

print("\nOriginal Data:\n")
print(data)
print("\nPreprocessed Data (stemmed & lemmatized):\n")
print(preprocessed_data)
```

```python
# LEARNING BAG OF WORDS
from sklearn.feature_extraction.text import CountVectorizer

# Use the preprocessed_data instead of the original 'data'
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(preprocessed_data)

print("\nDocuments (Preprocessed, Row Names):\n")
print(preprocessed_data)

print("\nVocabulary (Column Names/Feature Names):\n")
print(vectorizer.get_feature_names_out())

print("\nDocument-Term Matrix (X.toarray()):\n")
print(X.toarray())
```

```python
#LEARN TFIDF

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
x=vectorizer.fit_transform(preprocessed_data)

print("\nTF-IDF Vocabulary (Feature Names):\n")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Document-Term Matrix:\n")
print(x.toarray())
```

```python
tfidf_matrix = x.toarray()
feature_names = vectorizer.get_feature_names_out()

print("\nTF-IDF Matrix:\n")
print(tfidf_matrix)

print("\nFeature Names (Vocabulary):\n")
print(feature_names)
```

```python
import pandas as pd

tfidf_df = pd.DataFrame(tfidf_matrix, columns=feature_names, index=preprocessed_data)

print("\nTF-IDF DataFrame:\n")
print(tfidf_df.head())
```
