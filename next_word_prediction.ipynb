```python
import kagglehub


path = kagglehub.dataset_download("muhammadbilalhaneef/sherlock-holmes-next-word-prediction-corpus")

print("Path to dataset files:", path)

import os
print(os.listdir(path))
```

```python
file_path = os.path.join(path, 'Sherlock Holmes.txt')

with open(file_path, 'r', encoding='utf-8') as f:
    data = f.read()

print(f"Data loaded successfully. First 500 characters:\n{data[:500]}")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([data])
tokenizer.word_index
```

```python
input_seq=[]
for sentence in data.split('\n'):
    if sentence.strip():
        sequences = tokenizer.texts_to_sequences([sentence])
        if sequences and sequences[0]:
            # print(sequences[0])
            for i in range(1, len(sequences[0])):
                n_gram_sequence = sequences[0][:i+1]
                input_seq.append(n_gram_sequence)

for i in range(7):
    print(input_seq[i])
```

```python
input_seq
```

```python
max_sent=max([len(x) for x in input_seq])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_inp_seq=pad_sequences(input_seq,maxlen=max_sent,padding='pre')
```
